```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Car_price_prediction.csv')
df
```

```python
df=df.drop(["car_ID","symboling"],axis=1)
df
```

```python
car=df.select_dtypes(include=['int64','float64'])
```

```python
x=car.iloc[:,0:13].values
y=car.iloc[:,-1].values
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression as l
model=l()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
y_test
```

```python
df1=pd.DataFrame({'Actual':y_test, 'Predicted':y_predict})
df1
```

```python
df1.plot(kind='bar')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
